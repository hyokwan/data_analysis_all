### 아파트 매매 실거래 현황 데이터 수집 (* 법정동 코드 및 기간 반복문 추가해야 완성됨)

```python
import requests
from lxml import html
import bs4
# for dealYearMonth .. 반복
# for eachLawCode in range(lawCodeList): 반복:
aptTradeDataUrl = f"http://apis.data.go.kr/1613000/RTMSDataSvcAptTradeDev/getRTMSDataSvcAptTradeDev?serviceKey=AT33s775KYpJOkUBJu0dxkJuUeIfDIOJRzAH084EQS3JN%2BzFjErLHuk%2FGZa9L4gBTSGCzeA69tI9PwLp7B37IQ%3D%3D&LAWD_CD=11110&DEAL_YMD=202407&pageNo=1&numOfRows=10"
resp = requests.get(aptTradeDataUrl)
xmlObj = bs4.BeautifulSoup( resp.text, "lxml-xml")


rowList = []
columnList = []
columnNameList = []

itemTag = xmlObj.find(name="items")
itemList = itemTag.findAll(name="item")
for j in range(0, len(itemList)):
    #itemList 한개를 가지고 rowList 만들어보자!
    eachRow = itemList[j]
    allColumns = eachRow.findAll()
    # 전체컬럼 데이터를 columnList에 저장한다
    for i in range(0, len(allColumns)):
        eachColumnValue = allColumns[i].text
        columnList.append(eachColumnValue)
        # 첫번째 행을 돌때만 column정보를 수집해서 저장
        if j == 0:
            columnNameList.append( allColumns[i].name )
    rowList.append( columnList)
    columnList = []
aptTradeDf = pd.DataFrame( rowList, columns = columnNameList  )
aptTradeDf
# aptTradeDf = pd.DataFrame( rowList, columns = columnNameList  )
# aptTradeDf
```
