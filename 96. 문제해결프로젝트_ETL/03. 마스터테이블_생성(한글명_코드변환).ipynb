```python
import pandas as pd
```

### 테스트 데이터 셋 생성

```python
row1 = [ "미국","ST0001","202401",120]
row2 = [ "미국","ST0001","202402",90]
row3 = [ "한국","ST0001","202401",300]
row4 = [ "한국","ST0001","202402",20]
row5 = [ "중국","ST0001","202403",50]
```

```python
rawDf = pd.DataFrame( [row1, row2, row3, row4, row5],
                      columns = ["REGION","PRODUCT","YEARMONTH","SELLIN"])
```

```python
rawDf
```

### 마스터 테이블 생성

```python
regionMst = rawDf.REGION.drop_duplicates().reset_index(drop=True).reset_index()
```

```python
regionMst = regionMst.rename(columns={"index":"REGIONID"})
```

```python
preFix = "AA_"
```

```python
regionMst.REGIONID = preFix + regionMst.REGIONID.astype(str)
regionMst
```

### 업체 배포용 데이터셋 생성

```python
joinKey = "REGION"
openColumns = ["REGIONID","PRODUCT","YEARMONTH","SELLIN"]
mergedData = pd.merge( left  = rawDf, right = regionMst, how="left", on = joinKey)
companyDf = mergedData.loc[ :, openColumns ]
companyDf
```

### 업체에서 예측모델 생성

```python
companyDf["PREDICT"] = 100
companyDf
```

### 주관업체에서 데이터 활용 식별가능한 이름으로 변환

```python
joinKey = ["REGIONID"]
```

```python
finalDf = pd.merge( left = companyDf, right = regionMst,
          on=joinKey, how="left")
finalDf
```
