```python
# 스프레드시트의 행과열이 있는 데이터 조작 라이브러리
import pandas as pd
import numpy as np

# 머신러닝 모델 라이브러리
from sklearn import tree, ensemble, svm
from sklearn.tree import plot_tree

# 시각화 라이브러리
import matplotlib.pyplot as plt

# 머신러닝 전처리 라이브러리
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# 머신러닝 정확도 관련 라이브러리
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# 시간계산 라이브러리
from datetime import datetime
```

# 1. 타입 통합/ 특성숫자컬럼 추가

```python
ynDict = {"Y":1,"N":0}
featuresData["LE_PROMOTION"] = featuresData.PROMOTION.map( ynDict )
featuresData["LE_HOLIDAY"] = featuresData.HOLIDAY.map( ynDict )
```

# 2. 특성 선정 및 데이터 분리

```python
### 특성 선정
```

```python
# 시뮬레이션 시 아래 방법 활용
# simulCorrList = [0.3, 0.4, 0.5, 0.6, 0.7]
# for i in range(0, len(simulCorrList)):
```

```python
corrDf = featuresData.corr(method="pearson", numeric_only=True)
# 기준이 되는 상관계수를 어떻게 설정하느냐에 따라서
stdCorr = 0.3
label = ["QTY"]
features = list( corrDf.loc[ (abs( corrDf.QTY) >= stdCorr) &
                            (abs( corrDf.QTY) != 1)    ].index )
print( f"문제지 {features}, 정답지: {label} " )
### 데이터 분리 (시계열 데이터 접근하는 방법)
### 다양한 데이터 분리 방법
# 특정 시점으로 구분해서 시뮬레이션
splitRatio = 0.8
splitIndex = int( featuresData.shape[0] * splitRatio )
targetColumns = ["YEARWEEK"]
splitYearweek = featuresData.loc[ splitIndex, targetColumns ].values[0]

testData = featuresData.loc[ featuresData.YEARWEEK > splitYearweek ]
trainData = featuresData.loc[ featuresData.YEARWEEK <= splitYearweek ]
testData = featuresData.loc[ featuresData.YEARWEEK > splitYearweek ]

# TrainTestSplit 함수 활용 데이터 분리
# trainData, testData = train_test_split( featuresData, test_size = 0.2, random_state = 255)

trainingDataFeatures = trainData.loc[ :, features]
trainingDataLabel = trainData.loc[ :, label]
testDataFeatures = testData.loc[ :, features]
testDataLabel = testData.loc[ :, label]

print(trainingDataFeatures.shape)
print(trainingDataLabel.shape)
print(testDataFeatures.shape)
print(testDataLabel.shape)
### featuresData 연도로 구분한다  구분!
### TrainData = 2015 TestData = 2016
### trainingDataFeatures
### testDataFeatures
### trainingDataLabel
### testDataLabel
### 데이터 분리 (쉽게 접근하는 방법)
# trainData, testData = train_test_split( featuresData, test_size=0.2, random_state=1 )
# ## 훈련, 테스트데이터 내부에서 문제지 정답지 분리
# trainingDataFeatures = trainData.loc[ :, features]
# trainingDataLabel = trainData.loc[ :, label]
# testDataFeatures = testData.loc[ :, features]
# testDataLabel = testData.loc[ :, label]
# print(trainingDataFeatures.shape)
# print(trainingDataLabel.shape)
# print(testDataFeatures.shape)
# print(testDataLabel.shape)
```

### MAPE 활용 ACCURACY 계산

```python
#*** 1-ABS(FCST-ACT) / FCST *** QTY:40 FCST 20 (1/2이하 예측 시 마이너스) -> FCST * 2 <= QTY 0으로 설정
# 1-ABS(ACT-FCST) / ACT  *** QTY:40 FCST 80 (2배이상 예측 시 마이너스) -> FCST >= QTY * 2 0으로 설정
```

```python
testDataLabel["DT_ACCURACY"] = np.where( testDataLabel.PREDICT_DT * 2 <= testDataLabel.QTY, 0,
          1-abs(testDataLabel.PREDICT_DT -  testDataLabel.QTY) / testDataLabel.PREDICT_DT )
```

### 실습문제 2 풀이

```python
carsUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/cars.csv"
carsData = pd.read_csv(carsUrl)
carsData.head()

# 상관관계 분석 -> features 선발
corrDf = carsData.corr( method = "pearson", numeric_only = True)
stdCorr = 0.3
label = ["sales"]
features = list( corrDf.loc[ (abs( corrDf.sales ) >= stdCorr) &
                             (abs( corrDf.sales ) != 1)    ].index )

print( f"특성: {features}, 예측 목표: {label} " )

# +) 시각화로 상관관계 분석하여 features 선발함 제시!
import seaborn as sns
corrDf = carsData.corr( method = "pearson", numeric_only = True)
salesCorr = corrDf[["sales"]].sort_values( by = "sales", ascending = False )
sns.heatmap( data = salesCorr, annot=True )

# 훈련 / 테스트 데이터 분리하기 (2:8 분할!)
trainData, testData = train_test_split( carsData, test_size=0.2, random_state=1 )

trainDataFeatures = trainData.loc[ :, features]
trainDataLabel = trainData.loc[ :, label]
testDataFeatures = testData.loc[ :, features]
testDataLabel = testData.loc[ :, label]

# decision tree 훈련 모델 적용 + 예측
dtModel = tree.DecisionTreeRegressor( random_state= 1 )
fittedDtModel = dtModel.fit( trainDataFeatures, trainDataLabel )

dtPredictValue = fittedDtModel.predict( testDataFeatures )
testDataLabel["PREDICT_SALES"] = dtPredictValue
testDataLabel.head()

# 정확도 검증 (mean absolute percentage error 이용)
from sklearn.metrics import mean_absolute_percentage_error
mapeSales = mean_absolute_percentage_error( testDataLabel.sales, testDataLabel.PREDICT_SALES )
mapeSales * 100
```
