```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

### 1. 스케일링 (minmax)

```python
csData = pd.read_csv("https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/customer.csv")
csData
### minmax 스케일러
# 존재하는 함수 활용하여 스케일링 작업

balScaler = MinMaxScaler()
csData["scaler_balance"] = balScaler.fit_transform( csData.loc[ :, ["balance"]] )
# 커스터마이징 하기위한 스케일링 함수 생성
maxBalance = csData.balance.max()
minBalance = csData.balance.min()
def minMaxScler( inValue ):
    # inValue = 100
    bunmo = maxBalance - minBalance
    bunja = inValue - minBalance

    if bunmo == 0:
        return 0
    else:
        return bunja / bunmo
csData["bal_scaler2"] = csData.balance.apply (minMaxScler)
csData.describe()
```

### 2. 숫자변환

```python
# label 인코더 잘못 사용한 사례
csData = pd.read_csv("https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/customer.csv")
csData
csDict = {"normal":1, "diamond":2, "vip":3}
csData["le_label"] = csData.label.map( csDict )
csLabel = LabelEncoder()
csData["le_label2"] = csLabel.fit_transform( csData.label )
csLabel.classes_
csData.corr(numeric_only=True)

# label 인코더 외에 딕셔너리를 쫌 치밀하게 적용한 경우
groupKey = ["label"]
groupCsData = csData.groupby( groupKey )["balance"].agg(["mean"]).reset_index()
sortKey = ["mean"]
sortedData = groupCsData.sort_values( by = sortKey ).reset_index(drop=True)
labelOrder = list( sortedData.label )
csDict = {}
for i in range(0, len(labelOrder)):
    csDict[ labelOrder[i] ] = i
csDict
csData.corr(numeric_only=True)
```
